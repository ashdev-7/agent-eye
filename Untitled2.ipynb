Converting sequences to 3d arrays

```python
print("=" * 80)
print("STEP 5: WINDOWING - CREATE 3D SEQUENCES FOR RNN/LSTM")
print("=" * 80)

WINDOW_SIZE = 30

print(f"\n1️⃣  WINDOW CONFIGURATION")
print("-" * 80)
print(f"Window size: {WINDOW_SIZE} timesteps (cycles)")
print(f"Reason: Covers ~15% of typical 200-cycle engine lifecycle")

# Function to create windows
def create_windows(df, feature_cols, window_size=30):
    """
    Convert engine trajectories into sliding windows.

    Returns:
        X: (n_windows, window_size, n_features) - 3D array
        y: (n_windows, 1) - RUL values at window end
        metadata: dict with statistics
    """
    X_windows = []
    y_windows = []
    n_engines_processed = 0
    n_engines_skipped = 0

    for engine_id, group in df.groupby('engine_id'):
        group = group.sort_values('cycle')
        features = group[feature_cols].values  # (n_cycles, n_features)
        rul_values = group['rul'].values if 'rul' in group.columns else None

        n_cycles = len(features)

        # Skip engines with fewer cycles than window
        if n_cycles < window_size:
            n_engines_skipped += 1
            continue

        # Create sliding windows
        n_windows = n_cycles - window_size + 1
        for i in range(n_windows):
            window_features = features[i:i+window_size]  # (30, 22)
            X_windows.append(window_features)

            if rul_values is not None:
                y_windows.append(rul_values[i+window_size-1])  # RUL at window end

        n_engines_processed += 1

    X = np.array(X_windows)  # (n_windows, 30, 22)
    y = np.array(y_windows).reshape(-1, 1) if y_windows else None  # (n_windows, 1)

    metadata = {
        'n_engines_processed': n_engines_processed,
        'n_engines_skipped': n_engines_skipped,
        'n_windows': len(X_windows),
        'X_shape': X.shape,
        'y_shape': y.shape if y is not None else None
    }

    return X, y, metadata

# Create windows for TRAIN
print(f"\n2️⃣  CREATING TRAINING WINDOWS")
print("-" * 80)

X_train, y_train, train_meta = create_windows(train_df, feature_cols, WINDOW_SIZE)

print(f"Engines processed: {train_meta['n_engines_processed']}")
print(f"Engines skipped (<{WINDOW_SIZE} cycles): {train_meta['n_engines_skipped']}")
print(f"Total windows created: {train_meta['n_windows']}")
print(f"Window shape: {X_train.shape}")
print(f"RUL shape: {y_train.shape}")

# Create windows for TEST
print(f"\n3️⃣  CREATING TEST WINDOWS")
print("-" * 80)

# First, compute TEST RUL labels
def compute_test_rul(df, rul_ground_truth, cap=125):
    """
    Compute RUL for test data using ground truth file.
    Test engines are censored (run stopped early).
    RUL = (max_cycle_observed - current_cycle) + rul_from_file
    """
    y_test = np.zeros(len(df), dtype=np.float32)

    for idx, (engine_id, group) in enumerate(df.groupby('engine_id')):
        group = group.sort_values('cycle')
        max_cycle = group['cycle'].max()

        for local_idx, global_idx in enumerate(group.index):
            current_cycle = group.iloc[local_idx]['cycle']
            rul_value = (max_cycle - current_cycle) + rul_ground_truth[idx]
            y_test[global_idx] = np.clip(rul_value, 0, cap)

    return y_test

test_rul = compute_test_rul(test_df, rul_ground_truth, cap=125)
test_df['rul'] = test_rul

X_test, y_test, test_meta = create_windows(test_df, feature_cols, WINDOW_SIZE)

print(f"Engines processed: {test_meta['n_engines_processed']}")
print(f"Engines skipped (<{WINDOW_SIZE} cycles): {test_meta['n_engines_skipped']}")
print(f"Total windows created: {test_meta['n_windows']}")
print(f"Window shape: {X_test.shape}")
print(f"RUL shape: {y_test.shape}")

# Summary statistics
print(f"\n4️⃣  WINDOWED DATA STATISTICS")
print("-" * 80)

print(f"\nTraining Windows:")
print(f"   Total: {X_train.shape[0]:,}")
print(f"   RUL - Min: {y_train.min():.0f}, Max: {y_train.max():.0f}, Mean: {y_train.mean():.1f}")

print(f"\nTest Windows:")
print(f"   Total: {X_test.shape[0]:,}")
print(f"   RUL - Min: {y_test.min():.0f}, Max: {y_test.max():.0f}, Mean: {y_test.mean():.1f}")

print(f"\nData multiplication factor:")
print(f"   Train: 100 engines → {X_train.shape[0]:,} windows ({X_train.shape[0]/100:.0f}x)")
print(f"   Test: 100 engines → {X_test.shape[0]:,} windows ({X_test.shape[0]/100:.0f}x)")

# Save windowed data for later use
print(f"\n5️⃣  SAVING WINDOWED DATA")
print("-" * 80)

np.save('data/processed/X_train.npy', X_train)
np.save('data/processed/y_train.npy', y_train)
np.save('data/processed/X_test.npy', X_test)
np.save('data/processed/y_test.npy', y_test)

print(f"✓ X_train saved: data/processed/X_train.npy {X_train.shape}")
print(f"✓ y_train saved: data/processed/y_train.npy {y_train.shape}")
print(f"✓ X_test saved: data/processed/X_test.npy {X_test.shape}")
print(f"✓ y_test saved: data/processed/y_test.npy {y_test.shape}")

print("\n" + "=" * 80)
print("✅ WINDOWING COMPLETE - READY FOR MODEL TRAINING")
print("=" * 80)

print(f"\nNext step: Train baseline models (LSTM, GRU) and novel Dual Residual Attention model")
```

```python
print("=" * 80)
print("STEP 6: DEFINE EVALUATION METRICS")
print("=" * 80)

def compute_rmse(y_true, y_pred):
    """
    Root Mean Squared Error
    Formula: sqrt(mean((y_true - y_pred)^2))
    """
    return float(np.sqrt(np.mean((y_pred.flatten() - y_true.flatten())**2)))

def compute_nasa_score(y_true, y_pred):
    """
    NASA Asymmetric Scoring Function
    Late predictions (engine fails): exp(e/10) - 1  [HIGH PENALTY]
    Early predictions (false alarm): exp(-e/13) - 1 [LOWER PENALTY]
    where e = y_pred - y_true
    """
    e = y_pred.flatten() - y_true.flatten()
    late_penalty = np.where(e >= 0, np.exp(e / 10.0) - 1.0, 0.0)
    early_penalty = np.where(e < 0, np.exp(-e / 13.0) - 1.0, 0.0)
    return float(np.sum(late_penalty + early_penalty))

print("\n✓ RMSE function defined")
print("  Formula: sqrt(mean((y_true - y_pred)^2))")
print("  Interpretation: Lower is better. Units = cycles.")

print("\n✓ NASA Score function defined")
print("  Late error (e >= 0): exp(e/10) - 1")
print("  Early error (e < 0): exp(-e/13) - 1")
print("  Interpretation: Lower is better. Late penalties ~1.75x higher.")

# Test metrics on dummy data
y_dummy_true = np.array([50, 30, 10, 80, 125])
y_dummy_pred = np.array([55, 28, 12, 75, 120])

rmse_test = compute_rmse(y_dummy_true, y_dummy_pred)
nasa_test = compute_nasa_score(y_dummy_true, y_dummy_pred)

print(f"\nTest on dummy predictions:")
print(f"  True: {y_dummy_true}")
print(f"  Pred: {y_dummy_pred}")
print(f"  RMSE: {rmse_test:.2f}")
print(f"  NASA: {nasa_test:.2f}")

print("\n" + "=" * 80)
print("✅ METRICS READY")
print("=" * 80)
print(f"\nBenchmark targets (Wang et al. 2021):")
print(f"  RMSE: 12.31 cycles")
print(f"  NASA: 252 points")
print(f"\nOur targets:")
print(f"  LSTM baseline: ~13-14 RMSE, ~300-400 NASA")
print(f"  GRU baseline: ~13-14 RMSE, ~300-400 NASA")
print(f"  Dual Residual: ≤15 RMSE, ≤400 NASA (with interpretability bonus)")
```

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
import json
import pandas as pd
from scipy import stats

# Custom metric functions (implement as per your definitions)
def compute_rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred) ** 2))

def compute_nasa_score(y_true, y_pred):
    # NASA score weighted error metric (implementation depends on your legacy code)
    # Placeholder example:
    diff = y_pred - y_true
    score = np.sum(np.exp(-diff / 13) - 1)
    return score

# =====================================================================
# Build Dual Residual Attention LSTM Model
# =====================================================================
def build_dual_residual_attention_lstm(
    input_shape=(30, 22), encoder_units=192, attention_eps=0.25
):
    x_in = layers.Input(shape=input_shape, name='input')
    num_features = input_shape[-1]

    # Sensor Attention
    x_mean = layers.Lambda(lambda x: tf.reduce_mean(x, axis=1), name='sensor_mean')(x_in)
    a = layers.Dense(64, activation='tanh', name='sensor_hidden')(x_mean)
    alpha = layers.Dense(num_features, activation='softmax', name='sensor_attention')(a)
    alpha_expanded = layers.Lambda(lambda a: tf.expand_dims(a, axis=1))(alpha)

    def residual_gate(inputs):
        x, alpha_e = inputs
        F, eps = num_features, attention_eps
        gate = 1.0 + eps * (alpha_e - 1.0 / F)
        return x * gate

    x_gated = layers.Lambda(residual_gate, name='residual_gate')([x_in, alpha_expanded])

    # LSTM Encoder
    encoder = layers.LSTM(
        encoder_units,
        return_sequences=True,
        dropout=0.2,
        recurrent_dropout=0.2,
        name='lstm_encoder'
    )
    hidden_states = encoder(x_gated)
    hidden_states = layers.Dropout(0.2)(hidden_states)

    # Temporal Attention
    e_t = layers.Dense(1, activation='tanh', name='temporal_scores')(hidden_states)
    e_t_flat = layers.Flatten(name='temporal_scores_flat')(e_t)
    beta = layers.Activation('softmax', name='temporal_attention')(e_t_flat)
    beta_expanded = layers.Lambda(lambda b: tf.expand_dims(b, axis=-1))(beta)
    weighted_hidden = layers.Multiply()([hidden_states, beta_expanded])

    # Context aggregation
    context = layers.Lambda(lambda x: tf.reduce_sum(x, axis=1), name='context_aggregation')(weighted_hidden)

    # Prediction layers
    x = layers.Dense(64, activation='relu', name='dense_1')(context)
    x = layers.Dropout(0.2)(x)
    x = layers.Dense(32, activation='relu', name='dense_2')(x)
    x = layers.Dropout(0.2)(x)
    rul_output = layers.Dense(1, name='rul_prediction')(x)

    model = models.Model(inputs=x_in, outputs=[rul_output, alpha, beta], name='DualResidualAttention_LSTM')
    return model

dual_lstm_model = build_dual_residual_attention_lstm(input_shape=(30, 22), encoder_units=192)

# =====================================================================
# Prepare Data - Load and Split
# =====================================================================
X_train = np.load('data/processed/X_train.npy')
y_train = np.load('data/processed/y_train.npy')
X_test = np.load('data/processed/X_test.npy')
y_test = np.load('data/processed/y_test.npy')

from sklearn.model_selection import train_test_split

X_train_split, X_val, y_train_split, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42)

# =====================================================================
# Compile Model with Custom Loss Configuration
# =====================================================================
zero_sensor_train = np.zeros((y_train_split.shape[0], 22))
zero_temporal_train = np.zeros((y_train_split.shape[0], 30))
zero_sensor_val = np.zeros((y_val.shape[0], 22))
zero_temporal_val = np.zeros((y_val.shape[0], 30))

dual_lstm_model.compile(
    optimizer=optimizers.Adam(learning_rate=6e-4),
    loss={
        'rul_prediction': 'mse',
        'sensor_attention': lambda y_true, y_pred: 0.0,
        'temporal_attention': lambda y_true, y_pred: 0.0
    },
    loss_weights={
        'rul_prediction': 1.0,
        'sensor_attention': 0.0,
        'temporal_attention': 0.0
    },
    metrics={'rul_prediction': ['mae']}
)

early_stop = callbacks.EarlyStopping(
    monitor='val_rul_prediction_loss',
    patience=10,
    restore_best_weights=True,
    verbose=1,
    mode='min'  # Explicitly specify minimize mode for loss
)


# =====================================================================
# Training
# =====================================================================
history_dual = dual_lstm_model.fit(
    X_train_split,
    {'rul_prediction': y_train_split,
     'sensor_attention': zero_sensor_train,
     'temporal_attention': zero_temporal_train},
    validation_data=(
        X_val,
        {'rul_prediction': y_val,
         'sensor_attention': zero_sensor_val,
         'temporal_attention': zero_temporal_val}),
    epochs=60,
    batch_size=64,
    callbacks=[early_stop],
    verbose=1
)

# =====================================================================
# Prediction and Attention Extraction
# =====================================================================
original_model = models.Model(
    inputs=dual_lstm_model.input,
    outputs=[dual_lstm_model.get_layer('rul_prediction').output,
             dual_lstm_model.get_layer('sensor_attention').output,
             dual_lstm_model.get_layer('temporal_attention').output]
)

predictions_test = original_model.predict(X_test, verbose=0)
y_pred_lstm_att = predictions_test[0]
alpha_lstm_att = predictions_test[1]
beta_lstm_att = predictions_test[2]

print(f"Shapes - RUL: {y_pred_lstm_att.shape}, Sensor Attn: {alpha_lstm_att.shape}, Temporal Attn: {beta_lstm_att.shape}")

# =====================================================================
# Metric Calculation
# =====================================================================
rmse_lstm_att = compute_rmse(y_test, y_pred_lstm_att)
nasa_lstm_att = compute_nasa_score(y_test, y_pred_lstm_att)
mae_lstm_att = mean_absolute_error(y_test, y_pred_lstm_att)

print(f"LSTM Attention Test RMSE: {rmse_lstm_att:.2f}")
print(f"LSTM Attention Test NASA Score: {nasa_lstm_att:.0f}")
print(f"LSTM Attention Test MAE: {mae_lstm_att:.2f}")

# =====================================================================
# Create Directories For Outputs
# =====================================================================
os.makedirs('models', exist_ok=True)
os.makedirs('attention', exist_ok=True)
os.makedirs('figs', exist_ok=True)
os.makedirs('tables', exist_ok=True)

# =====================================================================
# Visualization: Sensor Attention
# =====================================================================
mean_alpha = np.mean(alpha_lstm_att, axis=0)
std_alpha = np.std(alpha_lstm_att, axis=0)
sensor_names = [f's{i}' for i in range(1, 23)]
sorted_indices = np.argsort(mean_alpha)[::-1]
sorted_alpha = mean_alpha[sorted_indices]
sorted_alpha_std = std_alpha[sorted_indices]
sorted_names = [sensor_names[i] for i in sorted_indices]

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes[0, 0].barh(range(10), sorted_alpha[:10], xerr=sorted_alpha_std[:10],
                color='steelblue', alpha=0.8, edgecolor='navy', linewidth=1.5, capsize=5)
axes[0, 0].set_yticks(range(10))
axes[0, 0].set_yticklabels(sorted_names[:10], fontsize=11, fontweight='bold')
axes[0, 0].set_xlabel('Mean Attention Weight', fontsize=11, fontweight='bold')
axes[0, 0].set_title('Top 10 Most Important Sensors (LSTM+Attn)', fontsize=12, fontweight='bold')
axes[0, 0].grid(True, alpha=0.3, axis='x')
for i, v in enumerate(sorted_alpha[:10]):
    axes[0, 0].text(v + 0.005, i, f'{v:.4f}', va='center', fontweight='bold')

axes[0, 1].bar(range(22), mean_alpha, yerr=std_alpha,
               color='coral', alpha=0.7, edgecolor='darkred', linewidth=1, capsize=3)
axes[0, 1].axhline(y=1/22, color='green', linestyle='--', linewidth=2, label='Uniform (0.0455)')
axes[0, 1].set_xticks(range(22))
axes[0, 1].set_xticklabels([f's{i}' for i in range(1, 23)], fontsize=9, rotation=45)
axes[0, 1].set_ylabel('Mean Attention Weight', fontsize=11, fontweight='bold')
axes[0, 1].set_title('All Sensors Importance (LSTM+Attn)', fontsize=12, fontweight='bold')
axes[0, 1].legend(fontsize=10)
axes[0, 1].grid(True, alpha=0.3, axis='y')

im = plt.imshow(alpha_lstm_att[:100, :], aspect='auto', cmap='RdYlGn', interpolation='nearest')
plt.colorbar(im, label='Attention Weight')
plt.xlabel('Sensor ID')
plt.ylabel('Test Sample')
plt.title('Sensor Attention Heatmap (First 100 Samples, LSTM+Attn)')
plt.savefig('figs/lstm_attention_sensor_importance_heatmap.png', dpi=100, bbox_inches='tight')
plt.close()
print("✓ Saved: figs/lstm_attention_sensor_importance_heatmap.png")

# =====================================================================
# Visualization: Temporal Attention
# =====================================================================
mean_beta = np.mean(beta_lstm_att, axis=0)
std_beta = np.std(beta_lstm_att, axis=0)

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes[0, 0].plot(range(1, 31), mean_beta, 'o-', linewidth=2.5, markersize=7, color='darkblue', label='Mean')
axes[0, 0].fill_between(range(1, 31), mean_beta - std_beta, mean_beta + std_beta, alpha=0.3, color='lightblue', label='±1 std')
axes[0, 0].set_xlabel('Timestep (Cycle)', fontsize=11, fontweight='bold')
axes[0, 0].set_ylabel('Attention Weight', fontsize=11, fontweight='bold')
axes[0, 0].set_title('Temporal Attention Over Time (LSTM+Attn)', fontsize=12, fontweight='bold')
axes[0, 0].legend(fontsize=10)
axes[0, 0].grid(True, alpha=0.3)

recent = np.mean(mean_beta[-10:])
middle = np.mean(mean_beta[10:20])
old = np.mean(mean_beta[:10])

categories = ['Cycles 1-10\n(Old)', 'Cycles 11-20\n(Middle)', 'Cycles 21-30\n(Recent)']
values = [old, middle, recent]
colors_bar = ['lightcoral', 'lightyellow', 'lightgreen']

bars = axes[0, 1].bar(categories, values, color=colors_bar, edgecolor='black', linewidth=2)
axes[0, 1].set_ylabel('Mean Attention', fontsize=11, fontweight='bold')
axes[0, 1].set_title('Old vs Recent Cycles (LSTM+Attn)', fontsize=12, fontweight='bold')
axes[0, 1].grid(True, alpha=0.3, axis='y')
for bar, val in zip(bars, values):
    height = bar.get_height()
    axes[0, 1].text(bar.get_x() + bar.get_width()/2., height, f'{val:.4f}', ha='center', va='bottom',
                    fontweight='bold', fontsize=11)

im = plt.imshow(beta_lstm_att[:100, :].T, aspect='auto', cmap='viridis', interpolation='nearest')
plt.colorbar(im, label='Attention Weight')
plt.ylabel('Timestep (Cycle)')
plt.xlabel('Test Sample')
plt.title('Temporal Attention Heatmap (First 100 Samples, LSTM+Attn)')
plt.savefig('figs/lstm_attention_temporal_importance_heatmap.png', dpi=100, bbox_inches='tight')
plt.close()
print("✓ Saved: figs/lstm_attention_temporal_importance_heatmap.png")

# =====================================================================
# Save Model and Artifacts
# =====================================================================
dual_lstm_model.save('models/dual_residual_attention_lstm.keras')
np.save('attention/alpha_lstm_att.npy', alpha_lstm_att)
np.save('attention/beta_lstm_att.npy', beta_lstm_att)

print("✓ Model and attention weights saved successfully.")

# =====================================================================
# Update and Save Comparison Table
# =====================================================================
comparison_data = {
    'Model': ['LSTM Baseline', 'GRU Baseline', 'Dual Residual Attention + LSTM', 'Dual Residual Attention + GRU'],
    'Encoder Units': [64, 64, 192, 192],
    'Test RMSE': [14.15, 14.14, rmse_lstm_att, rmse_dual_test],
    'Test NASA': [30752, 30466, nasa_lstm_att, nasa_dual_test],
    'Test MAE': [10.72, 11.03, mae_lstm_att, mae_dual_test]
}
comparison_df = pd.DataFrame(comparison_data)
comparison_df.to_csv('tables/model_comparison_v2.csv', index=False)
print("Updated comparison table saved at 'tables/model_comparison_v2.csv'")
print(comparison_df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import models

# Define extraction models for LSTM and GRU (adjust layer names as per your model)
original_lstm_model = models.Model(
    inputs=dual_lstm_model.input,
    outputs=[
        dual_lstm_model.get_layer('rul_prediction').output,
        dual_lstm_model.get_layer('sensor_attention').output,
        dual_lstm_model.get_layer('temporal_attention').output
    ]
)

original_gru_model = models.Model(
    inputs=dual_gru_model.input,
    outputs=[
        dual_gru_model.get_layer('rul_prediction').output,
        dual_gru_model.get_layer('sensor_attention').output,
        dual_gru_model.get_layer('temporal_attention').output
    ]
)

# Run predictions to extract attention weights (use your test data)
y_pred_lstm, alpha_lstm_att, beta_lstm_att = original_lstm_model.predict(X_test)
y_pred_gru, alpha_gru_att, beta_gru_att = original_gru_model.predict(X_test)

sensor_names = [f's{i}' for i in range(1, alpha_lstm_att.shape[1]+1)]
timesteps = range(1, beta_lstm_att.shape[1]+1)

def plot_sensor_attention(alpha, model_name):
    mean_alpha = np.mean(alpha, axis=0)
    std_alpha = np.std(alpha, axis=0)
    sorted_indices = np.argsort(mean_alpha)[::-1]
    sorted_alpha = mean_alpha[sorted_indices]
    sorted_alpha_std = std_alpha[sorted_indices]
    sorted_names = [sensor_names[i] for i in sorted_indices]

    plt.figure(figsize=(10,6))
    plt.barh(sorted_names[:10], sorted_alpha[:10], xerr=sorted_alpha_std[:10], color='steelblue')
    plt.xlabel('Mean Sensor Attention Weight')
    plt.title(f'Top 10 Sensors Importance - {model_name}')
    plt.gca().invert_yaxis()
    plt.show()

def plot_temporal_attention(beta, model_name):
    mean_beta = np.mean(beta, axis=0)
    std_beta = np.std(beta, axis=0)

    plt.figure(figsize=(12,6))
    plt.plot(timesteps, mean_beta, label='Mean Attention', marker='o')
    plt.fill_between(timesteps, mean_beta - std_beta, mean_beta + std_beta, alpha=0.3, label='±1 std')
    plt.xlabel('Timestep (Cycle)')
    plt.ylabel('Temporal Attention Weight')
    plt.title(f'Temporal Attention Over Time - {model_name}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Visualize LSTM Attention
plot_sensor_attention(alpha_lstm_att, "LSTM + Attention")
plot_temporal_attention(beta_lstm_att, "LSTM + Attention")

# Visualize GRU Attention
plot_sensor_attention(alpha_gru_att, "GRU + Attention")
plot_temporal_attention(beta_gru_att, "GRU + Attention")
```
