Converting sequences to 3d arrays

```python
print("=" * 80)
print("STEP 5: WINDOWING - CREATE 3D SEQUENCES FOR RNN/LSTM")
print("=" * 80)

WINDOW_SIZE = 30

print(f"\n1️⃣  WINDOW CONFIGURATION")
print("-" * 80)
print(f"Window size: {WINDOW_SIZE} timesteps (cycles)")
print(f"Reason: Covers ~15% of typical 200-cycle engine lifecycle")

# Function to create windows
def create_windows(df, feature_cols, window_size=30):
    """
    Convert engine trajectories into sliding windows.

    Returns:
        X: (n_windows, window_size, n_features) - 3D array
        y: (n_windows, 1) - RUL values at window end
        metadata: dict with statistics
    """
    X_windows = []
    y_windows = []
    n_engines_processed = 0
    n_engines_skipped = 0

    for engine_id, group in df.groupby('engine_id'):
        group = group.sort_values('cycle')
        features = group[feature_cols].values  # (n_cycles, n_features)
        rul_values = group['rul'].values if 'rul' in group.columns else None

        n_cycles = len(features)

        # Skip engines with fewer cycles than window
        if n_cycles < window_size:
            n_engines_skipped += 1
            continue

        # Create sliding windows
        n_windows = n_cycles - window_size + 1
        for i in range(n_windows):
            window_features = features[i:i+window_size]  # (30, 22)
            X_windows.append(window_features)

            if rul_values is not None:
                y_windows.append(rul_values[i+window_size-1])  # RUL at window end

        n_engines_processed += 1

    X = np.array(X_windows)  # (n_windows, 30, 22)
    y = np.array(y_windows).reshape(-1, 1) if y_windows else None  # (n_windows, 1)

    metadata = {
        'n_engines_processed': n_engines_processed,
        'n_engines_skipped': n_engines_skipped,
        'n_windows': len(X_windows),
        'X_shape': X.shape,
        'y_shape': y.shape if y is not None else None
    }

    return X, y, metadata

# Create windows for TRAIN
print(f"\n2️⃣  CREATING TRAINING WINDOWS")
print("-" * 80)

X_train, y_train, train_meta = create_windows(train_df, feature_cols, WINDOW_SIZE)

print(f"Engines processed: {train_meta['n_engines_processed']}")
print(f"Engines skipped (<{WINDOW_SIZE} cycles): {train_meta['n_engines_skipped']}")
print(f"Total windows created: {train_meta['n_windows']}")
print(f"Window shape: {X_train.shape}")
print(f"RUL shape: {y_train.shape}")

# Create windows for TEST
print(f"\n3️⃣  CREATING TEST WINDOWS")
print("-" * 80)

# First, compute TEST RUL labels
def compute_test_rul(df, rul_ground_truth, cap=125):
    """
    Compute RUL for test data using ground truth file.
    Test engines are censored (run stopped early).
    RUL = (max_cycle_observed - current_cycle) + rul_from_file
    """
    y_test = np.zeros(len(df), dtype=np.float32)

    for idx, (engine_id, group) in enumerate(df.groupby('engine_id')):
        group = group.sort_values('cycle')
        max_cycle = group['cycle'].max()

        for local_idx, global_idx in enumerate(group.index):
            current_cycle = group.iloc[local_idx]['cycle']
            rul_value = (max_cycle - current_cycle) + rul_ground_truth[idx]
            y_test[global_idx] = np.clip(rul_value, 0, cap)

    return y_test

test_rul = compute_test_rul(test_df, rul_ground_truth, cap=125)
test_df['rul'] = test_rul

X_test, y_test, test_meta = create_windows(test_df, feature_cols, WINDOW_SIZE)

print(f"Engines processed: {test_meta['n_engines_processed']}")
print(f"Engines skipped (<{WINDOW_SIZE} cycles): {test_meta['n_engines_skipped']}")
print(f"Total windows created: {test_meta['n_windows']}")
print(f"Window shape: {X_test.shape}")
print(f"RUL shape: {y_test.shape}")

# Summary statistics
print(f"\n4️⃣  WINDOWED DATA STATISTICS")
print("-" * 80)

print(f"\nTraining Windows:")
print(f"   Total: {X_train.shape[0]:,}")
print(f"   RUL - Min: {y_train.min():.0f}, Max: {y_train.max():.0f}, Mean: {y_train.mean():.1f}")

print(f"\nTest Windows:")
print(f"   Total: {X_test.shape[0]:,}")
print(f"   RUL - Min: {y_test.min():.0f}, Max: {y_test.max():.0f}, Mean: {y_test.mean():.1f}")

print(f"\nData multiplication factor:")
print(f"   Train: 100 engines → {X_train.shape[0]:,} windows ({X_train.shape[0]/100:.0f}x)")
print(f"   Test: 100 engines → {X_test.shape[0]:,} windows ({X_test.shape[0]/100:.0f}x)")

# Save windowed data for later use
print(f"\n5️⃣  SAVING WINDOWED DATA")
print("-" * 80)

np.save('data/processed/X_train.npy', X_train)
np.save('data/processed/y_train.npy', y_train)
np.save('data/processed/X_test.npy', X_test)
np.save('data/processed/y_test.npy', y_test)

print(f"✓ X_train saved: data/processed/X_train.npy {X_train.shape}")
print(f"✓ y_train saved: data/processed/y_train.npy {y_train.shape}")
print(f"✓ X_test saved: data/processed/X_test.npy {X_test.shape}")
print(f"✓ y_test saved: data/processed/y_test.npy {y_test.shape}")

print("\n" + "=" * 80)
print("✅ WINDOWING COMPLETE - READY FOR MODEL TRAINING")
print("=" * 80)

print(f"\nNext step: Train baseline models (LSTM, GRU) and novel Dual Residual Attention model")
```

```python
print("=" * 80)
print("STEP 7: BASELINE MODEL 1 - SIMPLE LSTM")
print("=" * 80)

import numpy as np

def compute_rmse(y_true: np.ndarray, y_pred: np.ndarray) -> float:
    """
    Compute Root Mean Squared Error between true and predicted values.
    """
    return np.sqrt(np.mean((y_true - y_pred) ** 2))

def compute_nasa_score(y_true: np.ndarray, y_pred: np.ndarray, alpha: float=13.0, beta: float=10.0) -> float:
    """
    Compute NASA score for RUL prediction error:
    https://ti.arc.nasa.gov/tech/dash/groups/pcoe/prognostic-data-repository/
    """
    diffs = y_pred.flatten() - y_true.flatten()
    score = 0.
    for e in diffs:
        if e < 0:
            score += np.exp(-e / alpha) - 1
        else:
            score += np.exp(e / beta) - 1
    return score



# Reload windowed data if needed
X_train = np.load('data/processed/X_train.npy')
y_train = np.load('data/processed/y_train.npy')
X_test = np.load('data/processed/X_test.npy')
y_test = np.load('data/processed/y_test.npy')

print(f"\nData loaded:")
print(f"  X_train: {X_train.shape}")
print(f"  y_train: {y_train.shape}")
print(f"  X_test: {X_test.shape}")
print(f"  y_test: {y_test.shape}")

# Split train into train/val (80/20)
from sklearn.model_selection import train_test_split
X_train_split, X_val, y_train_split, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42
)

print(f"\nTrain-Val split (80-20):")
print(f"  X_train_split: {X_train_split.shape}")
print(f"  X_val: {X_val.shape}")

# Build LSTM model
print("\n1️⃣  BUILDING LSTM ARCHITECTURE")
print("-" * 80)

input_shape = (X_train.shape[1], X_train.shape[2])  # (30, 22)

lstm_model = models.Sequential([
    layers.LSTM(64, input_shape=input_shape, return_sequences=False),
    layers.Dropout(0.2),
    layers.Dense(32, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(1)
])

print(f"Model created:")
print(f"  Input shape: {input_shape}")
print(f"  Architecture: LSTM(64) → Dropout → Dense(32) → Dense(1)")

lstm_model.summary()

# Compile
lstm_model.compile(
    optimizer=optimizers.Adam(learning_rate=1e-3),
    loss='mse',
    metrics=['mae']
)

print(f"\n✓ Model compiled")
print(f"  Optimizer: Adam (LR=0.001)")
print(f"  Loss: MSE (mean squared error)")
print(f"  Metrics: MAE (mean absolute error)")

# Train with early stopping
print(f"\n2️⃣  TRAINING LSTM")
print("-" * 80)

early_stop = callbacks.EarlyStopping(
    monitor='val_loss',
    patience=8,
    restore_best_weights=True,
    verbose=1
)

start_time = time.time()

print(f"Starting training...")
history_lstm = lstm_model.fit(
    X_train_split, y_train_split,
    validation_data=(X_val, y_val),
    epochs=50,
    batch_size=64,
    callbacks=[early_stop],
    verbose=1
)

training_time_lstm = time.time() - start_time
epochs_lstm = len(history_lstm.history['loss'])

print(f"\n✓ Training complete")
print(f"  Total time: {training_time_lstm:.1f} seconds")
print(f"  Epochs trained: {epochs_lstm}")
print(f"  Final train loss: {history_lstm.history['loss'][-1]:.4f}")
print(f"  Final val loss: {history_lstm.history['val_loss'][-1]:.4f}")

# Plot training history
plt.figure(figsize=(10, 4))
plt.plot(history_lstm.history['loss'], label='Train Loss', linewidth=2)
plt.plot(history_lstm.history['val_loss'], label='Val Loss', linewidth=2)
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.title('LSTM Training History')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('figs/lstm_training_history.png', dpi=100, bbox_inches='tight')
plt.close()
print(f"\n✓ Training history plot saved: figs/lstm_training_history.png")

# Evaluate on test set
print(f"\n3️⃣  EVALUATING LSTM ON TEST SET")
print("-" * 80)

y_pred_lstm_train = lstm_model.predict(X_train_split, verbose=0)
y_pred_lstm_test = lstm_model.predict(X_test, verbose=0)

rmse_lstm_train = compute_rmse(y_train_split, y_pred_lstm_train)
rmse_lstm_test = compute_rmse(y_test, y_pred_lstm_test)
nasa_lstm_test = compute_nasa_score(y_test, y_pred_lstm_test)
mae_lstm_test = mean_absolute_error(y_test, y_pred_lstm_test)

print(f"\nLSTM Results:")
print(f"  Train RMSE: {rmse_lstm_train:.2f} cycles")
print(f"  Test RMSE:  {rmse_lstm_test:.2f} cycles")
print(f"  Test NASA:  {nasa_lstm_test:.0f} points")
print(f"  Test MAE:   {mae_lstm_test:.2f} cycles")

# Compare to Wang benchmark
print(f"\nComparison to Wang et al. 2021:")
print(f"  Wang RMSE: 12.31 cycles")
print(f"  Your RMSE: {rmse_lstm_test:.2f} cycles")
print(f"  Difference: {rmse_lstm_test - 12.31:+.2f} cycles ({100*(rmse_lstm_test-12.31)/12.31:+.1f}%)")

print(f"\n  Wang NASA: 252 points")
print(f"  Your NASA: {nasa_lstm_test:.0f} points")
print(f"  Difference: {nasa_lstm_test - 252:+.0f} points ({100*(nasa_lstm_test-252)/252:+.1f}%)")

# Plot predictions vs ground truth
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.scatter(y_test, y_pred_lstm_test, alpha=0.5, s=10)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2, label='Perfect prediction')
plt.xlabel('True RUL')
plt.ylabel('Predicted RUL')
plt.title(f'LSTM Predictions (RMSE={rmse_lstm_test:.2f})')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
errors = y_pred_lstm_test.flatten() - y_test.flatten()
plt.hist(errors, bins=50, edgecolor='black', alpha=0.7)
plt.xlabel('Prediction Error (cycles)')
plt.ylabel('Frequency')
plt.title('Error Distribution')
plt.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('figs/lstm_predictions.png', dpi=100, bbox_inches='tight')
plt.close()
print(f"\n✓ Prediction plots saved: figs/lstm_predictions.png")



# Save model
lstm_model.save('models/lstm_baseline.h5')
print(f"\n✓ Model saved: models/lstm_baseline.h5")

print("\n" + "=" * 80)
print("✅ LSTM BASELINE COMPLETE")
print("=" * 80)
```

```python
print("\n" + "=" * 120)
print("4️⃣  TRAINING DUAL RESIDUAL ATTENTION MODEL (RESEARCH GRADE)")
print("=" * 120)

# ============================================================================
# BUILD DUAL RESIDUAL ATTENTION GRU MODEL
# ============================================================================

def build_dual_residual_attention_gru(input_shape=(30, 22), encoder_units=192, attention_eps=0.25):
    """
    Build dual residual attention GRU model with sensor and temporal attention.
    Research-grade implementation with proper tensor operations.
    """
    x_in = layers.Input(shape=input_shape, name='input')
    num_features = input_shape[-1]
    num_timesteps = input_shape[0]

    # SENSOR ATTENTION BRANCH
    x_mean = layers.GlobalAveragePooling1D()(x_in)
    a = layers.Dense(64, activation='tanh', name='sensor_hidden')(x_mean)
    alpha = layers.Dense(num_features, activation='softmax', name='sensor_attention')(a)
    alpha_expanded = layers.Reshape((1, num_features))(alpha)

    gate_coeff = layers.Lambda(
        lambda a: 1.0 - attention_eps + attention_eps * a,
        output_shape=(1, num_features),
        name='gate_coefficient'
    )(alpha_expanded)

    x_gated = layers.Multiply(name='residual_gate')([x_in, gate_coeff])

    # GRU ENCODER + TEMPORAL ATTENTION
    hidden_states = layers.GRU(
        encoder_units,
        return_sequences=True,
        dropout=0.2,
        recurrent_dropout=0.2,
        name='gru_encoder'
    )(x_gated)

    hidden_states = layers.Dropout(0.2)(hidden_states)

    # TEMPORAL ATTENTION MECHANISM
    et = layers.Dense(1, activation='tanh')(hidden_states)
    et_flat = layers.Flatten()(et)
    beta = layers.Activation('softmax', name='temporal_attention')(et_flat)
    beta_expanded = layers.Reshape((num_timesteps, 1))(beta)

    weighted_hidden = layers.Multiply()([hidden_states, beta_expanded])

    context = layers.Lambda(
        lambda x: tf.reduce_sum(x, axis=1),
        output_shape=(encoder_units,),
        name='temporal_context'
    )(weighted_hidden)

    # RUL PREDICTION HEAD
    x = layers.Dense(64, activation='relu', name='dense1')(context)
    x = layers.Dropout(0.2)(x)
    x = layers.Dense(32, activation='relu', name='dense2')(x)
    x = layers.Dropout(0.2)(x)
    rul_output = layers.Dense(1, name='rul_prediction')(x)

    # CREATE MODEL WITH 3 OUTPUTS
    model = models.Model(
        inputs=x_in,
        outputs=[rul_output, alpha, beta],
        name='DualResidualAttentionGRU'
    )

    return model

# Create the full model (3 outputs for attention extraction)
dual_gru_full = build_dual_residual_attention_gru(
    input_shape=(30, 22),
    encoder_units=192,
    attention_eps=0.25
)

print("✓ Dual Residual Attention GRU model created")
print(dual_gru_full.summary())

# ============================================================================
# REBUILD MODEL TO ONLY OUTPUT RUL FOR TRAINING
# ============================================================================

dual_gru_model = models.Model(
    inputs=dual_gru_full.input,
    outputs=dual_gru_full.get_layer('rul_prediction').output
)

dual_gru_model.compile(
    optimizer=optimizers.Adam(learning_rate=6e-4),
    loss='mse',
    metrics=['mae']
)

print("\n✓ Single-output model created and compiled for training")

# ============================================================================
# LOAD DATA
# ============================================================================

X_train = np.load('data/processed/X_train.npy')
y_train = np.load('data/processed/y_train.npy')
X_test = np.load('data/processed/X_test.npy')
y_test = np.load('data/processed/y_test.npy')

X_train_split, X_val, y_train_split, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42
)

print(f"✓ Data loaded:")
print(f"  X_train_split: {X_train_split.shape}, y_train_split: {y_train_split.shape}")
print(f"  X_val: {X_val.shape}, y_val: {y_val.shape}")
print(f"  X_test: {X_test.shape}, y_test: {y_test.shape}")

# ============================================================================
# TRAINING
# ============================================================================

early_stop = callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True,
    verbose=1
)

print("\n" + "=" * 120)
print("Starting training...")
print(f"  Max epochs: 60")
print(f"  Batch size: 64")
print(f"  Early stopping patience: 10 epochs")
print(f"  Learning rate: 0.0006")
print("=" * 120)

start_time = time.time()

history_dual = dual_gru_model.fit(
    X_train_split,
    y_train_split,
    validation_data=(X_val, y_val),
    epochs=60,
    batch_size=64,
    callbacks=[early_stop],
    verbose=1
)

training_time_dual = time.time() - start_time
epochs_trained_dual = len(history_dual.history['loss'])

# Extract best epoch
best_epoch_dual = history_dual.history['val_loss'].index(min(history_dual.history['val_loss'])) + 1

print(f"\n✓ Training complete")
print(f"   Total time: {training_time_dual:.1f} seconds")
print(f"   Epochs trained: {epochs_trained_dual}")
print(f"   Best epoch: {best_epoch_dual}")
print(f"   Final train loss: {history_dual.history['loss'][-1]:.4f}")
print(f"   Final val loss: {history_dual.history['val_loss'][-1]:.4f}")

# ============================================================================
# EXTRACT PREDICTIONS & ATTENTION WEIGHTS
# ============================================================================

print("\n" + "=" * 120)
print("5️⃣  EXTRACTING PREDICTIONS & ATTENTION WEIGHTS")
print("=" * 120)

# Use the full 3-output model for prediction
predictions_train = dual_gru_full.predict(X_train_split, verbose=0)
y_pred_dual_train = predictions_train[0]
alpha_train = predictions_train[1]
beta_train = predictions_train[2]

predictions_test = dual_gru_full.predict(X_test, verbose=0)
y_pred_dual_test = predictions_test[0]
alpha_test = predictions_test[1]
beta_test = predictions_test[2]

print(f"✓ Predictions extracted:")
print(f"  Train RUL: {y_pred_dual_train.shape}")
print(f"  Test RUL: {y_pred_dual_test.shape}")
print(f"  Test Sensor Attention: {alpha_test.shape}")
print(f"  Test Temporal Attention: {beta_test.shape}")

# Compute metrics
rmse_dual_train = compute_rmse(y_train_split, y_pred_dual_train)
rmse_dual_test = compute_rmse(y_test, y_pred_dual_test)
nasa_dual_test = compute_nasa_score(y_test, y_pred_dual_test)
mae_dual_test = mean_absolute_error(y_test, y_pred_dual_test)

print(f"\n✓ DUAL RESIDUAL ATTENTION + GRU RESULTS:")
print(f"   Train RMSE: {rmse_dual_train:.2f} cycles")
print(f"   Test RMSE:  {rmse_dual_test:.2f} cycles")
print(f"   Test NASA:  {nasa_dual_test:.0f} points")
print(f"   Test MAE:   {mae_dual_test:.2f} cycles")

# ============================================================================
# REPRODUCIBILITY LOGGING
# ============================================================================

print("\n" + "=" * 120)
print("6️⃣  LOGGING RESULTS FOR REPRODUCIBILITY")
print("=" * 120)

result_row = {
    "Model": "GRU_DualAttention",
    "Test_RMSE": round(rmse_dual_test, 4),
    "Test_NASA": round(nasa_dual_test, 2),
    "Test_MAE": round(mae_dual_test, 4),
    "Epochs": best_epoch_dual,
    "Seed": 42
}

results_csv = "tables/model_comparison.csv"
df = pd.DataFrame([result_row])
if os.path.exists(results_csv):
    df_old = pd.read_csv(results_csv)
    df = pd.concat([df_old, df], ignore_index=True)
df.to_csv(results_csv, index=False)

print(f"✓ Results logged to {results_csv}")
print(f"  RMSE: {rmse_dual_test:.4f}, NASA: {nasa_dual_test:.0f}, MAE: {mae_dual_test:.4f}")

# ============================================================================
# SAVE MODEL AND ATTENTION WEIGHTS
# ============================================================================

import os
os.makedirs('models', exist_ok=True)
os.makedirs('attention', exist_ok=True)

# Save model
dual_gru_model.save('models/GRU_DualAttention_42.h5')
print(f"\n✓ Model saved: models/GRU_DualAttention_42.h5")

# Save attention weights
np.save('attention/alpha_test_gru.npy', alpha_test)
np.save('attention/beta_test_gru.npy', beta_test)
print(f"✓ Attention weights saved:")
print(f"  - attention/alpha_test_gru.npy (sensor attention)")
print(f"  - attention/beta_test_gru.npy (temporal attention)")

print("\n" + "=" * 120)
print("✅ DUAL RESIDUAL ATTENTION + GRU TRAINING COMPLETE")
print("=" * 120)
```

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
import json
import pandas as pd
from scipy import stats

# Custom metric functions (implement as per your definitions)
def compute_rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred) ** 2))

def compute_nasa_score(y_true, y_pred):
    # NASA score weighted error metric (implementation depends on your legacy code)
    # Placeholder example:
    diff = y_pred - y_true
    score = np.sum(np.exp(-diff / 13) - 1)
    return score

# =====================================================================
# Build Dual Residual Attention LSTM Model
# =====================================================================
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks

def build_dual_residual_attention_lstm(
    input_shape=(30, 22),
    encoder_units=192,
    attention_eps=0.25,
    seed=42
):
    tf.random.set_seed(seed)
    tf.keras.utils.set_random_seed(seed)

    x_in = layers.Input(shape=input_shape, name='input')
    num_features = input_shape[-1]

    x_mean = layers.Lambda(lambda x: tf.reduce_mean(x, axis=1), name='sensor_mean')(x_in)

    a = layers.Dense(
        64, activation='tanh', name='sensor_hidden',
        kernel_initializer=tf.keras.initializers.GlorotUniform(seed=seed)
    )(x_mean)

    alpha = layers.Dense(
        num_features, activation='softmax', name='sensor_attention',
        kernel_initializer=tf.keras.initializers.GlorotUniform(seed=seed)
    )(a)

    alpha_expanded = layers.Lambda(lambda a: tf.expand_dims(a, axis=1))(alpha)

    def residual_gate(inputs):
        x, alpha_e = inputs
        F, eps = num_features, attention_eps
        gate = 1.0 + eps * (alpha_e - 1.0 / F)
        return x * gate

    x_gated = layers.Lambda(residual_gate, name='residual_gate')([x_in, alpha_expanded])

    hidden_states = layers.LSTM(
        encoder_units,
        return_sequences=True,
        dropout=0.2,
        recurrent_dropout=0.2,
        name='lstm_encoder',
        kernel_initializer=tf.keras.initializers.GlorotUniform(seed=seed),
        recurrent_initializer=tf.keras.initializers.Orthogonal(seed=seed),
        bias_initializer='zeros'
    )(x_gated)

    hidden_states = layers.Dropout(0.2)(hidden_states)

    e_t = layers.Dense(
        1, activation='tanh', name='temporal_scores',
        kernel_initializer=tf.keras.initializers.GlorotUniform(seed=seed)
    )(hidden_states)

    e_t_flat = layers.Flatten(name='temporal_scores_flat')(e_t)
    beta = layers.Activation('softmax', name='temporal_attention')(e_t_flat)
    beta_expanded = layers.Lambda(lambda b: tf.expand_dims(b, axis=-1))(beta)
    weighted_hidden = layers.Multiply()([hidden_states, beta_expanded])

    context = layers.Lambda(lambda x: tf.reduce_sum(x, axis=1), name='context_aggregation')(weighted_hidden)

    x = layers.Dense(
        64, activation='relu', name='dense_1',
        kernel_initializer=tf.keras.initializers.GlorotUniform(seed=seed)
    )(context)
    x = layers.Dropout(0.2)(x)

    x = layers.Dense(
        32, activation='relu', name='dense_2',
        kernel_initializer=tf.keras.initializers.GlorotUniform(seed=seed)
    )(x)
    x = layers.Dropout(0.2)(x)

    rul_output = layers.Dense(
        1, name='rul_prediction',
        kernel_initializer=tf.keras.initializers.GlorotUniform(seed=seed)
    )(x)

    model = models.Model(inputs=x_in, outputs=[rul_output, alpha, beta], name='DualResidualAttention_LSTM')
    return model


# ✅ UPDATE THIS LINE
dual_lstm_model = build_dual_residual_attention_lstm(
    input_shape=(30, 22),
    encoder_units=192,
    seed=SEED  # ← ADD THIS
)


dual_lstm_model = build_dual_residual_attention_lstm(input_shape=(30, 22), encoder_units=192)

# =====================================================================
# Prepare Data - Load and Split
# =====================================================================
X_train = np.load('data/processed/X_train.npy')
y_train = np.load('data/processed/y_train.npy')
X_test = np.load('data/processed/X_test.npy')
y_test = np.load('data/processed/y_test.npy')

from sklearn.model_selection import train_test_split

X_train_split, X_val, y_train_split, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42)

# =====================================================================
# Compile Model with Custom Loss Configuration
# =====================================================================
zero_sensor_train = np.zeros((y_train_split.shape[0], 22))
zero_temporal_train = np.zeros((y_train_split.shape[0], 30))
zero_sensor_val = np.zeros((y_val.shape[0], 22))
zero_temporal_val = np.zeros((y_val.shape[0], 30))

dual_lstm_model.compile(
    optimizer=optimizers.Adam(learning_rate=6e-4),
    loss={
        'rul_prediction': 'mse',
        'sensor_attention': lambda y_true, y_pred: 0.0,
        'temporal_attention': lambda y_true, y_pred: 0.0
    },
    loss_weights={
        'rul_prediction': 1.0,
        'sensor_attention': 0.0,
        'temporal_attention': 0.0
    },
    metrics={'rul_prediction': ['mae']}
)

early_stop = callbacks.EarlyStopping(
    monitor='val_rul_prediction_loss',
    patience=10,
    restore_best_weights=True,
    verbose=1,
    mode='min'  # Explicitly specify minimize mode for loss
)


# =====================================================================
# Training
# =====================================================================
history_dual = dual_lstm_model.fit(
    X_train_split,
    {'rul_prediction': y_train_split,
     'sensor_attention': zero_sensor_train,
     'temporal_attention': zero_temporal_train},
    validation_data=(
        X_val,
        {'rul_prediction': y_val,
         'sensor_attention': zero_sensor_val,
         'temporal_attention': zero_temporal_val}),
    epochs=60,
    batch_size=64,
    callbacks=[early_stop],
    verbose=1
)

# =====================================================================
# Prediction and Attention Extraction
# =====================================================================
original_model = models.Model(
    inputs=dual_lstm_model.input,
    outputs=[dual_lstm_model.get_layer('rul_prediction').output,
             dual_lstm_model.get_layer('sensor_attention').output,
             dual_lstm_model.get_layer('temporal_attention').output]
)

predictions_test = original_model.predict(X_test, verbose=0)
y_pred_lstm_att = predictions_test[0]
alpha_lstm_att = predictions_test[1]
beta_lstm_att = predictions_test[2]

print(f"Shapes - RUL: {y_pred_lstm_att.shape}, Sensor Attn: {alpha_lstm_att.shape}, Temporal Attn: {beta_lstm_att.shape}")

# =====================================================================
# Metric Calculation
# =====================================================================
rmse_lstm_att = compute_rmse(y_test, y_pred_lstm_att)
nasa_lstm_att = compute_nasa_score(y_test, y_pred_lstm_att)
mae_lstm_att = mean_absolute_error(y_test, y_pred_lstm_att)

print(f"LSTM Attention Test RMSE: {rmse_lstm_att:.2f}")
print(f"LSTM Attention Test NASA Score: {nasa_lstm_att:.0f}")
print(f"LSTM Attention Test MAE: {mae_lstm_att:.2f}")

# =====================================================================
# Create Directories For Outputs
# =====================================================================
os.makedirs('models', exist_ok=True)
os.makedirs('attention', exist_ok=True)
os.makedirs('figs', exist_ok=True)
os.makedirs('tables', exist_ok=True)

# =====================================================================
# Visualization: Sensor Attention
# =====================================================================
mean_alpha = np.mean(alpha_lstm_att, axis=0)
std_alpha = np.std(alpha_lstm_att, axis=0)
sensor_names = [f's{i}' for i in range(1, 23)]
sorted_indices = np.argsort(mean_alpha)[::-1]
sorted_alpha = mean_alpha[sorted_indices]
sorted_alpha_std = std_alpha[sorted_indices]
sorted_names = [sensor_names[i] for i in sorted_indices]

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes[0, 0].barh(range(10), sorted_alpha[:10], xerr=sorted_alpha_std[:10],
                color='steelblue', alpha=0.8, edgecolor='navy', linewidth=1.5, capsize=5)
axes[0, 0].set_yticks(range(10))
axes[0, 0].set_yticklabels(sorted_names[:10], fontsize=11, fontweight='bold')
axes[0, 0].set_xlabel('Mean Attention Weight', fontsize=11, fontweight='bold')
axes[0, 0].set_title('Top 10 Most Important Sensors (LSTM+Attn)', fontsize=12, fontweight='bold')
axes[0, 0].grid(True, alpha=0.3, axis='x')
for i, v in enumerate(sorted_alpha[:10]):
    axes[0, 0].text(v + 0.005, i, f'{v:.4f}', va='center', fontweight='bold')

axes[0, 1].bar(range(22), mean_alpha, yerr=std_alpha,
               color='coral', alpha=0.7, edgecolor='darkred', linewidth=1, capsize=3)
axes[0, 1].axhline(y=1/22, color='green', linestyle='--', linewidth=2, label='Uniform (0.0455)')
axes[0, 1].set_xticks(range(22))
axes[0, 1].set_xticklabels([f's{i}' for i in range(1, 23)], fontsize=9, rotation=45)
axes[0, 1].set_ylabel('Mean Attention Weight', fontsize=11, fontweight='bold')
axes[0, 1].set_title('All Sensors Importance (LSTM+Attn)', fontsize=12, fontweight='bold')
axes[0, 1].legend(fontsize=10)
axes[0, 1].grid(True, alpha=0.3, axis='y')

im = plt.imshow(alpha_lstm_att[:100, :], aspect='auto', cmap='RdYlGn', interpolation='nearest')
plt.colorbar(im, label='Attention Weight')
plt.xlabel('Sensor ID')
plt.ylabel('Test Sample')
plt.title('Sensor Attention Heatmap (First 100 Samples, LSTM+Attn)')
plt.savefig('figs/lstm_attention_sensor_importance_heatmap.png', dpi=100, bbox_inches='tight')
plt.close()
print("✓ Saved: figs/lstm_attention_sensor_importance_heatmap.png")

# =====================================================================
# Visualization: Temporal Attention
# =====================================================================
mean_beta = np.mean(beta_lstm_att, axis=0)
std_beta = np.std(beta_lstm_att, axis=0)

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes[0, 0].plot(range(1, 31), mean_beta, 'o-', linewidth=2.5, markersize=7, color='darkblue', label='Mean')
axes[0, 0].fill_between(range(1, 31), mean_beta - std_beta, mean_beta + std_beta, alpha=0.3, color='lightblue', label='±1 std')
axes[0, 0].set_xlabel('Timestep (Cycle)', fontsize=11, fontweight='bold')
axes[0, 0].set_ylabel('Attention Weight', fontsize=11, fontweight='bold')
axes[0, 0].set_title('Temporal Attention Over Time (LSTM+Attn)', fontsize=12, fontweight='bold')
axes[0, 0].legend(fontsize=10)
axes[0, 0].grid(True, alpha=0.3)

recent = np.mean(mean_beta[-10:])
middle = np.mean(mean_beta[10:20])
old = np.mean(mean_beta[:10])

categories = ['Cycles 1-10\n(Old)', 'Cycles 11-20\n(Middle)', 'Cycles 21-30\n(Recent)']
values = [old, middle, recent]
colors_bar = ['lightcoral', 'lightyellow', 'lightgreen']

bars = axes[0, 1].bar(categories, values, color=colors_bar, edgecolor='black', linewidth=2)
axes[0, 1].set_ylabel('Mean Attention', fontsize=11, fontweight='bold')
axes[0, 1].set_title('Old vs Recent Cycles (LSTM+Attn)', fontsize=12, fontweight='bold')
axes[0, 1].grid(True, alpha=0.3, axis='y')
for bar, val in zip(bars, values):
    height = bar.get_height()
    axes[0, 1].text(bar.get_x() + bar.get_width()/2., height, f'{val:.4f}', ha='center', va='bottom',
                    fontweight='bold', fontsize=11)

im = plt.imshow(beta_lstm_att[:100, :].T, aspect='auto', cmap='viridis', interpolation='nearest')
plt.colorbar(im, label='Attention Weight')
plt.ylabel('Timestep (Cycle)')
plt.xlabel('Test Sample')
plt.title('Temporal Attention Heatmap (First 100 Samples, LSTM+Attn)')
plt.savefig('figs/lstm_attention_temporal_importance_heatmap.png', dpi=100, bbox_inches='tight')
plt.close()
print("✓ Saved: figs/lstm_attention_temporal_importance_heatmap.png")

# =====================================================================
# Save Model and Artifacts
# =====================================================================
dual_lstm_model.save('models/dual_residual_attention_lstm.keras')
np.save('attention/alpha_lstm_att.npy', alpha_lstm_att)
np.save('attention/beta_lstm_att.npy', beta_lstm_att)

print("✓ Model and attention weights saved successfully.")

# =====================================================================
# Update and Save Comparison Table
# =====================================================================
comparison_data = {
    'Model': ['LSTM Baseline', 'GRU Baseline', 'Dual Residual Attention + LSTM', 'Dual Residual Attention + GRU'],
    'Encoder Units': [64, 64, 192, 192],
    'Test RMSE': [14.15, 14.14, rmse_lstm_att, rmse_dual_test],
    'Test NASA': [30752, 30466, nasa_lstm_att, nasa_dual_test],
    'Test MAE': [10.72, 11.03, mae_lstm_att, mae_dual_test]
}
comparison_df = pd.DataFrame(comparison_data)
comparison_df.to_csv('tables/model_comparison_v2.csv', index=False)
print("Updated comparison table saved at 'tables/model_comparison_v2.csv'")
print(comparison_df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import models

# Define extraction models for LSTM and GRU (layer names must match your architecture)
def build_extraction_model(base_model):
    return models.Model(
        inputs=base_model.input,
        outputs=[
            base_model.get_layer('rul_prediction').output,
            base_model.get_layer('sensor_attention').output,
            base_model.get_layer('temporal_attention').output
        ]
    )

original_lstm_model = build_extraction_model(dual_lstm_model)
original_gru_model  = build_extraction_model(dual_gru_model)

# Predict to extract attention weights
y_pred_lstm, alpha_lstm, beta_lstm = original_lstm_model.predict(X_test)
y_pred_gru,  alpha_gru,  beta_gru  = original_gru_model.predict(X_test)

sensor_names = [f's{i}' for i in range(1, alpha_lstm.shape[1]+1)]
timesteps = np.arange(1, beta_lstm.shape[1]+1)

def plot_sensor_attention(alpha, model_name, sensor_names, n_top=10, save_path=None):
    mean_alpha = np.mean(alpha, axis=0)
    std_alpha = np.std(alpha, axis=0)
    sorted_indices = np.argsort(mean_alpha)[::-1]
    sorted_names  = [sensor_names[i] for i in sorted_indices[:n_top]]
    sorted_alpha  = mean_alpha[sorted_indices[:n_top]]
    sorted_alpha_std = std_alpha[sorted_indices[:n_top]]

    plt.figure(figsize=(10,6))
    plt.barh(sorted_names, sorted_alpha, xerr=sorted_alpha_std, color='steelblue', capsize=5)
    plt.xlabel('Mean Sensor Attention Weight')
    plt.title(f'Top {n_top} Sensors Importance - {model_name}')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=100)
    plt.show()

def plot_temporal_attention(beta, model_name, timesteps, save_path=None):
    mean_beta = np.mean(beta, axis=0)
    std_beta  = np.std(beta, axis=0)
    plt.figure(figsize=(12,6))
    plt.plot(timesteps, mean_beta, label='Mean Attention', marker='o')
    plt.fill_between(timesteps, mean_beta - std_beta, mean_beta + std_beta, alpha=0.3, label='±1 std')
    plt.xlabel('Timestep (Cycle)')
    plt.ylabel('Temporal Attention Weight')
    plt.title(f'Temporal Attention Over Time - {model_name}')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=100)
    plt.show()

# Save figures as well as displaying
plot_sensor_attention(alpha_lstm, "LSTM + Attention", sensor_names, save_path='figs/lstm_sensor_attention.png')
plot_temporal_attention(beta_lstm, "LSTM + Attention", timesteps, save_path='figs/lstm_temporal_attention.png')
plot_sensor_attention(alpha_gru,  "GRU + Attention",  sensor_names, save_path='figs/gru_sensor_attention.png')
plot_temporal_attention(beta_gru, "GRU + Attention",  timesteps, save_path='figs/gru_temporal_attention.png')
```

## Dual Attention Interpretation: LSTM & GRU Models

This section extracts and visualizes the sensor and temporal attention weights learned by both LSTM-DualAttention and GRU-DualAttention models. These visualizations provide critical insights into *how* the models leverage input data to make predictions for Remaining Useful Life (RUL).

### Sensor Attention Barplot

- **Plot:** Horizontal barplot ranking the top-10 most important sensors (features) for the prediction task.
- **Interpretation:**
    - *Bar Length*: Represents the average attention assigned to each sensor across all test samples; longer bars indicate greater importance.
    - *Error Bars*: Show the standard deviation, i.e., how consistently a sensor is attended to across the test set.
- **Insights:** Identifies which physical sensors have the strongest predictive power in the learned model, aiding feature engineering and sensor selection for further experimental design or deployment.

### Temporal Attention Timeseries

- **Plot:** Line plot showing average temporal attention across all input cycles/timesteps, with shaded regions for ±1 standard deviation.
- **Interpretation:**
    - *Peaks in attention*: Indicate input cycles/timesteps that most heavily influence the model’s prediction.
    - *Shaded area*: Reveals the stability or variability of attention at each timestep.
- **Insights:** Shows at which points in the input sequence the model ‘looks’ most attentively, revealing dynamic patterns useful for understanding failure progression or optimizing sequence windowing.

### Side-by-Side Model Comparison

By visualizing both attention mechanisms for LSTM and GRU architectures, we can contrast their interpretability and decide which approach captures the problem structure or operational priorities better.
